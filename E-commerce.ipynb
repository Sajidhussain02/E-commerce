```python
summary = df[["Quantity","UnitPrice","TotalAmount"]].agg(["mean","median","max","min","std"])
print (summary)
```

```python
top_products = df.groupby("ProductCategory")["TotalAmount"].sum().sort_values(ascending= False).head(5)
print(top_products)
```

```python
df["Quantity"].agg(["mean","min","max"])
```

```python
df["PurchaseDate"] = pd.to_datetime (df["PurchaseDate"])
```

```python
df['Month'] = df['PurchaseDate'].dt.to_period('M')
```

```python
monthly_sales = df.groupby("Month")["TotalAmount"].sum().reset_index()

plt.plot(monthly_sales['Month'].astype(str), monthly_sales['TotalAmount'], marker='o')
plt.xticks(rotation=45)  
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
payment_sales = df.groupby("PaymentMethod")["TotalAmount"].sum()
```

```python
payment_sales.plot.pie(autopct = "%1.1f%%", figsize=(5,5), title = "sales by payment method(pie)")
plt.show()
```

```python
payment_sales.plot.bar( color = "skyblue", title = "Sales by payment method") 
plt.ylabel("Total Sales")
plt.xlabel("Payment Method")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
           
```
